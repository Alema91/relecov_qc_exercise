```python
protocolo_diag = df[["ID","Sample ID","Protocolo diagnostico SARS-CoV-2"]]
```

```python
pd_df= protocolo_diag.loc[(protocolo_diag['Sample ID'] == '#1')]

```

```python
pd_df.drop(columns = 'Sample ID', inplace =True)
```

```python
pd_df['Protocolo diagnostico SARS-CoV-2'].replace("Vircell Multiplex", "RT-PCR, Vircell Multiplex",inplace = True)
```

```python
pd_df['Protocolo diagnostico SARS-CoV-2'].value_counts()
```

```python


pd_df.drop(axis=0, index=120, inplace=True)
```

```python
a_rep = pd_df.groupby(["Protocolo diagnostico SARS-CoV-2"]).count()
```

```python
d = {'Protocol': ["RT-PCR","Sequencing"], 'ID': [24,15]}
data_new = pd.DataFrame(data=d)
```

```python
data_new
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.pyplot import figure


# Fixing random state for reproducibility


#plt.rcdefaults()
fig, ax = plt.subplots()

fig.set_size_inches(6, 6)
# Example data
protocol = a_rep.index
y_pos = np.arange(len(protocol))
number = a_rep["ID"]

ax.barh(y_pos, number, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(protocol,fontsize = 7)
ax.bar_label(ax.containers[0],fontsize=10)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of laboratories', fontsize = 10)
ax.set_title('Diagnositc protocol', fontsize = 10)
plt.show()



fig.savefig('diagnostic_protocol_all.png',bbox_inches='tight')   # save the figure to file
plt.close(fig)    # close the figure window


```
